```python
#codice main
#questo codice genera il grafico
%matplotlib widget
import matplotlib.pyplot as plt
from IPython.display import display, Markdown, HTML
import pandas as pd
import json
import ipywidgets as widgets

plt.ion()

def get_coordinates(file):
    x = []
    y = []
    file = open(file, "r")
    jsonraw = json.load(file)
    jsonraw = jsonraw["node"]
    for year in jsonraw["years"]:
        x.append(year["year"])
    for data in jsonraw["datas"]:
        y.append(data["data"])
    return x, y

def get_calc(req_year):
    coun = 0
    toreet = {
        0: None,
        1: None,
        2: None,
        3: None,
    }
    files = ["final-production-v2104.json", "final-price-v2404.json", "final-yield-v2304.json", "final-harvestedarea-v2304.json"]
    for file in files:
        file_open = open(file, "r")
        jsonraw = json.load(file_open)
        jsonraw = jsonraw["node"]
        for year in jsonraw["years"]:
            if not toreet[coun]:
                if year["name"] == req_year:
                    for data in jsonraw["datas"]:
                        if data["id"] == year["id"]:
                            toreet[coun] = data["data"]
        coun += 1
    return toreet

display(HTML('''<h1 align="center">Dati sul grano</h1>
'''))

display(HTML('''<h3 align="center">Calcolatore per anno</h3>
'''))

calculator_year = widgets.Text(
    value='',
    placeholder='',
    description='Anno:',
    disabled=False
)
calculator_btn = widgets.Button(
    description='clicca',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='calcola',
    icon=''
)
calculator_result = widgets.HTML(
    value="",
    placeholder='',
    description='',
)

calculator_container = widgets.Box([calculator_year, calculator_btn, calculator_result])
display(calculator_container)

display(HTML('''<h3 align="center">Grafici</h3>
'''))
production_toggle = widgets.Checkbox(
    value=True,
    description='Produzione',
    disabled=False
)
price_toggle = widgets.Checkbox(
    value=True,
    description='Prezzo',
    disabled=False
)
yield_toggle = widgets.Checkbox(
    value=True,
    description='Raccolto',
    disabled=False
)
harvestedarea_toggle = widgets.Checkbox(
    value=True,
    description='Area coltivata',
    disabled=False
)
toggles_container = widgets.Box([production_toggle, price_toggle, yield_toggle, harvestedarea_toggle])
display(toggles_container)


production_years, production_datas= get_coordinates("final-production-v2104.json")
production_fig, production_ax = plt.subplots()
production_graph = production_ax.plot(production_years, production_datas,marker=".", markevery=[73,100])[0]
production_fig.canvas.header_visible = False
production_ax.set_title("PRODUZIONE")
production_ax.set_xlabel("anni")
production_ax.set_ylabel("milioni di tonnellate")
plt.text(1938, 165, "1")
plt.text(1945, 86.4, "2")
display(HTML('''<p align="center">1 - inizio 2° guerra mondiale<br />2 - fine 2° guerra mondiale</p>
'''))

price_years, price_datas = get_coordinates("final-price-v2404.json")
price_fig, price_ax = plt.subplots()
price_graph = price_ax.plot(price_years, price_datas)[0]
price_fig.canvas.header_visible = False
price_ax.set_title("PREZZO")
price_ax.set_xlabel("anni")
price_ax.set_ylabel("USD/tonnellate")
display(HTML('''<p align="center">!!! i dati non tegono conto della svalutazione</p>
'''))

yield_years, yield_datas = get_coordinates("final-yield-v2304.json")
yield_fig, yield_ax = plt.subplots()
yield_graph = yield_ax.plot(yield_years, yield_datas)[0]
yield_fig.canvas.header_visible = False
yield_ax.set_title("RACCOLTO")
yield_ax.set_xlabel("anni")
yield_ax.set_ylabel("ettogrammi/acri")

harvestedarea_years, harvestedarea_datas = get_coordinates("final-harvestedarea-v2304.json")
harvestedarea_fig, harvestedarea_ax = plt.subplots()
harvestedarea_graph = harvestedarea_ax.plot(harvestedarea_years, harvestedarea_datas)[0]
harvestedarea_fig.canvas.header_visible = False
harvestedarea_ax.set_title("SUPERFICIE COLTIVATA")
harvestedarea_ax.set_xlabel("anni")
harvestedarea_ax.set_ylabel("acri")

def handle_prod(e):
    if e["new"] == True:
        production_fig.canvas.layout.display = "inline-flex"
    elif e["new"] == False:
        production_fig.canvas.layout.display = "none"
production_toggle.observe(handle_prod, names="value")

def handle_price(e):
    if e["new"] == True:
        price_fig.canvas.layout.display = "inline-flex"
    elif e["new"] == False:
        price_fig.canvas.layout.display = "none"
price_toggle.observe(handle_price, names="value")

def handle_yield(e):
    if e["new"] == True:
        yield_fig.canvas.layout.display = "inline-flex"
    elif e["new"] == False:
        yield_fig.canvas.layout.display = "none"
yield_toggle.observe(handle_yield, names="value")

def handle_harvestedarea(e):
    if e["new"] == True:
        harvestedarea_fig.canvas.layout.display = "inline-flex"
    elif e["new"] == False:
        harvestedarea_fig.canvas.layout.display = "none"
harvestedarea_toggle.observe(handle_harvestedarea, names="value")

def handle_calculate(e):
    res = get_calc(calculator_year.value)
    if res[0]:
        res0 = res[0]
    else:
        res0 = "0"
    if res[1]:
        res1 = res[1]
    else:
        res1 = "0"
    if res[2]:
        res2 = res[2]
    else:
        res2 = "0"
    if res[3]:
        res3 = res[3]
    else:
        res3 = "0"
    prod_scacchi = ""
    if res0 != "0":
        prod_scacchi = ((1.195328909*(10**12))/float(res0))
    price_scacchi = ""
    if res1 != "0":
        price_scacchi =  ((1.195328909*(10**12))*float(res1))
    calculator_result.value = ""
    calculator_result.value = f'''<p>Produzione: {res0}T<br/>Prezzo: {res1}$/T<br/>Raccolto {res2}HG/HA<br /> Superficie coltivata: {res3}HA
    <br /> Numero di produzioni necessarie per riempire la scacchiera: {prod_scacchi}
    <br /> Costo del grano sulla scacchiera: {price_scacchi}$
    </p>'''

calculator_btn.on_click(handle_calculate)


```
